```python
import sympy 

def combinations_by_subset(seq, r):
    if r:
        for i in range(r - 1, len(seq)):
            for cl in combinations_by_subset(seq[:i], r - 1):
                yield cl + (seq[i],)
    else:
        yield tuple()

cc= list(sympy.primerange(0,10000000))
print(len(cc))
x=combinations_by_subset(cc,4)
for i in range(0,20):
    print(next(x))


```
